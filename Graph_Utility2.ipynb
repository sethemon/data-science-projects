## Degree of a Vertex

Each node of a graph has an in-degree and an out-degree. The in-degree is the number of incoming edges. The out-degree is the number of edges starting at this node (outcoming). The node is called a source if it has 0 in-degree. The node is called a leaf if it has 0 out-degree

## Eigenvector Centrality
In graph theory, eigenvector centrality (also called eigencentrality) is a measure of the influence of a node in a network. It assigns relative scores to all nodes in the network based on the concept that connections to high-scoring nodes contribute more to the score of the node in question than equal connections to low-scoring nodes.

```python
nx.eigenvector_centrality(G_symmetric)
```

## Closeness centrality for nodes.

Closeness centrality of a node u is the reciprocal of the sum of the shortest path distances from u to all n-1 other nodes. Since the sum of distances depends on the number of nodes in the graph, closeness is normalized by the sum of minimum possible distances n-1.

```python
nx.closeness_centrality(G_symmetric)
```

## Kartz  Centrality
In graph theory, the Katz centrality of a node is a measure of centrality in a network. Unlike typical centrality measures which consider only the shortest path  between a pair of nodes, Katz centrality measures influence by taking into account the total number of walks between a pair of actors.

It is similar to Googles PageRank and to the eigenvector centrality.

```python
nx.katz_centrality(G_symmetric)
```

```python
G_symmetric.nodes
```

## Group Betweeness Centrality

Group betweenness centrality of a group of nodes $C$ is the sum of the
fraction of all-pairs shortest paths that pass through any vertex in $C$

```python
nx.group_betweenness_centrality(G_symmetric,['Amitabh Bachchan'])
```

## group_closeness_centrality

Group closeness centrality of a group of nodes $S$ is a measure
of how close the group is to the other nodes in the graph.

```python
nx.group_closeness_centrality(G_symmetric,['Dev Anand'])
```

## Degree Histogram

```python
import collections
import matplotlib.pyplot as plt
```

```python
G = nx.gnp_random_graph(100, 0.2)
```

```python
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)  # degree sequence
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())
```

```python
deg
```

```python
cnt
```

## Indegree outdegree distribution

```python
cam_net = nx.read_edgelist('./../webinar-2/cambridge_net.txt',
create_using=nx.DiGraph(), nodetype=int)
N, K = cam_net.order(), cam_net.size()
avg_deg = float(K) / N
```

```python
in_degrees = cam_net.in_degree() # dictionary node:degree
in_values = sorted(set([val for (node, val) in in_degrees]))
in_hist = [[val for (node, val) in in_degrees].count(x) for x in in_values]

out_degrees = cam_net.in_degree() # dictionary node:degree
out_values = sorted(set([val for (node, val) in out_degrees]))
out_hist = [[val for (node, val) in out_degrees].count(x) for x in out_values]
```

```python
plt.figure() # you need to first do 'import pylab as plt'
plt.grid(True)
plt.plot(in_values, in_hist, 'ro-') # in-degree
plt.plot(out_values, out_hist, 'bv-') # out-degree
plt.legend(['In-degree', 'Out-degree'])
plt.xlabel('Degree')
plt.ylabel('Number of nodes')
plt.title('network of places in Cambridge')
plt.xlim([0, 2*10**2])
plt.show()
```
